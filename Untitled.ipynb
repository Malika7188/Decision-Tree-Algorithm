```python
from ucimlrepo import fetch_ucirepo
import pandas as pd

# Load dataset
dataset = fetch_ucirepo(id=383)  # Cervical Cancer (Risk Factors) dataset

# Separate features and target
X = dataset.data.features
y = dataset.data.targets

# Combine and save to CSV
df = pd.concat([X, y], axis=1)
df.to_csv('data/risk_factors_cervical_cancer.csv', index=False)

print("✅ Dataset saved to data/risk_factors_cervical_cancer.csv")
```
